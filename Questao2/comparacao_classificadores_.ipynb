# Teste de Friedman

```python
import numpy as np
```

```python
#0 - normal-multivariada
#1 - knn
#2 - parzen
#3 - regressao logistica
#4 - ensemble

#No nosso caso, como existe apenas um conjunto de dados, r = R, em que r eh o ranking e R eh o ranking medio

Rankings = [3.5,3.25,1.25,5,2]
```

```python
N = 4
k = 5

X_F_square = (12*N/(k*(k+1)))*(sum(np.square(Rankings)) - (k*((k+1)**2)/4))

X_F_square
```

```python
F_F = ((N-1)*X_F_square)/(N*(k-1) - X_F_square)

F_F
```

dale.png

Com 4 e 12 graus de liberdade e escolhendo alfa = 5%, o F critico vale 3,2592.

H0 deve ser rejeitada se F_F for maior que o valor critico.

F_F apresenta um valor superior ao valor critico e a hipotese nula deve ser rejeitada.
